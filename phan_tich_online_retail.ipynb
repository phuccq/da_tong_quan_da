```python
import pandas as pd
import numpy as np

data = pd.read_csv('OnlineRetail.csv')
data
```

```python
data.info()
```

```python
# Công ty bán hàng do bao nhiêu nước sản xuất
country_num=data['Country'].unique().size
print(f'Công ty bán hàng do {country_num} nước sản xuất')

# Tổng số lượng đơn hàng bán ra, tổng doanh thu
invoice_total = data['InvoiceNo'].unique().size
data['total'] = data['Quantity']*data['UnitPrice']
total = data['total'].sum()
print(f'Tổng số lượng đơn hàng bán ra là {invoice_total}, tổng doanh thu là {total:2.0f}')
```

```python
# Top 10 mặt hàng có số lượng bán ra lớn rất
data.groupby('Description')['Quantity'].sum().sort_values('Quantity',ascending=False).head(10)
```

```python
# Top 10 mặt hàng có doanh thu lớn nhất
data.groupby('Description')['total'].sum()
```
